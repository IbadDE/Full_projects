```python
import pandas as pd
from bs4 import BeautifulSoup as bs
import requests
```

```python
                                                # 2016 bowling

url = 'https://stats.espncricinfo.com/ci/engine/records/averages/bowling.html?id=10555;type=tournament'
source = requests.get(url)
soup = bs(source.text, 'html.parser')
player_profile = soup.find('tbody')
print(len(player_profile))



player_16 = []
team_16 = []

for player in player_profile.find_all('tr', class_ = 'data2'):
    name = player.find_all('td')[0].text
    matches = player.find_all('td')[1].text
    innings = player.find_all('td')[2].text
    overs = player.find_all('td')[3].text
    maidens = player.find_all('td')[4].text
    runs = player.find_all('td')[5].text
    wickets = player.find_all('td')[6].text
    best_bowling = player.find_all('td')[7].text
    average = player.find_all('td')[8].text
    economy = player.find_all('td')[9].text
    strike_rate = player.find_all('td')[10].text
    four_wkts = player.find_all('td')[11].text
    five_wkts = player.find_all('td')[12].text
    catches = player.find_all('td')[13].text
    stumps = player.find_all('td')[14].text
    season = '2015/2016'

    player_16.append((name, matches,innings,overs,maidens,runs,wickets,best_bowling,average,economy,strike_rate,four_wkts,five_wkts,catches,stumps,season))


for i in player_profile.find_all('tr', class_ = 'note'):
    team = i.find('td').text
    
    team_16.append(team)
    
    
df_1 = pd.DataFrame(player_16)
df_2 = pd.DataFrame(team_16)

df_16 = pd.concat([df_1, df_2], axis = 1)
print(df_16.shape)
df_16.head()
```
